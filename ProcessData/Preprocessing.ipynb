```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("../Scrapy/rever.com_spider/data/canho.json")
```

```python
df.info()
```

```python
rows_with_nan_in_title = df[df['title'].isna()]

# In ra các hàng có giá trị NaN trong cột 'title'
print("Các hàng có giá trị NaN trong cột 'title':")
print(rows_with_nan_in_title)

df['title'] = df['title'].fillna('KXĐ')
```

```python
duplicate_rows = df[df.duplicated(subset=['title'], keep='first')]

print("\nCác hàng trùng lặp sau lần xuất hiện đầu tiên:")
print(duplicate_rows.index)

df = df.drop_duplicates(subset=['title'], keep='first')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df.drop(['title','url'],axis=1)
```

```python
law_doc_values = df['law_doc'].unique()
furniture_values = df['furniture'].unique()
type_values = df['type'].unique()
```

```python
law_doc_mapping = { 'Sổ hồng': 0, 'HĐ mua bán': 1, 'Sổ đỏ': 2, 'Giấy tay': 3, 'KXĐ': -1}
furniture_mapping = {'Không có nội thất': 0, 'Đầy đủ nội thất' : 1, 'Nội thất cơ bản': 2, 'KXĐ': -1}

df['law_doc'] = df['law_doc'].map(law_doc_mapping)
df['furniture'] = df['furniture'].map(furniture_mapping)


df.head()
```

```python
def convert_price(price):
    price = price.strip()  # Loại bỏ khoảng trắng dư thừa
    if 'tỷ' in price:
        return float(price.replace(' tỷ', ''))  # Chuyển đổi "tỷ" thành số và nhân với 1 tỷ
    elif 'triệu' in price:
        return float(price.replace(' triệu', '')) * 0.001  # Chuyển đổi "triệu" thành số và nhân với 1 triệu
    elif ' tr' in price:
        return float(price.replace(' tr', '')) * 0.001
    else:
        return None  # Trường hợp giá trị không hợp lệ
    
def convert_price_per_area(price):
    price = price.strip()  # Loại bỏ khoảng trắng dư thừa nếu có
    if 'triệu/m²' in price:
        return float(price.replace(' triệu/m²', ''))  # Xử lý trường hợp "triệu/m²"
    elif 'nghìn/m²' in price:
        return float(price.replace(' nghìn/m²', '')) * 0.001  # Xử lý trường hợp "nghìn/m²", chuyển sang "triệu/m²"
    else:
        return None  # Trường hợp không hợp lệ hoặc giá trị không chứa "triệu" hoặc "nghìn"

    
df['price'] = df['price'].apply(convert_price)

df['price_per_area'] = df['price_per_area'].apply(convert_price_per_area)

df['area'] = df['area'].str.replace(' m²', '', regex=False).astype(float)


df.head()
```

```python
df['utilities_list'] = df['utilities'].fillna('').str.split(', ')

# Áp dụng hàm lambda để tạo dummy columns cho các tiện ích
utilities_dummies = df['utilities_list'].apply(lambda x: pd.Series(1, index=x)).fillna(0)

# Gộp các cột tiện ích vào DataFrame gốc
df = pd.concat([df, utilities_dummies], axis=1)

df.head()
```

```python
df = df.drop(['utilities','utilities_list', 'false', ""], axis = 1)
df.info()
```

```python
df['post_date'] = pd.to_datetime(df['post_date'], format='%d/%m/%Y')
df['post_date'] = df['post_date'].astype(str)
```
